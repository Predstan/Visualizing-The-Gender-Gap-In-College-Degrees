# Visualizing The Gender Gap In College Degrees

Introduction: In this Project, we will be Communicating the nuanced narrative of gender gap using effective data visualization The Department of Education Statistics releases a statistical data collection detailing the number of bachelor degrees granted to women from 1970 to 2012. The data set is split into 17 grade levels, each column being a different group. Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. 

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
women_degrees.head()
```

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
spines = ["bottom", "right", "left", "top"]
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
```

```python
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize = (18, 20))

for maj in range(0,6):
        ax = fig.add_subplot(6, 3, ((maj*3)+1))
        ax.plot(women_degrees["Year"], women_degrees[stem_cats[maj]], c = cb_dark_blue, label = "Women", linewidth =3)
        ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[maj]], c = cb_orange, label = "Men", linewidth = 3)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0, 100)
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "off")
        ax.set_title(stem_cats[maj])
        ax.set_yticks([0,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        for spot in spines:
            ax.spines[spot].set_visible(False)
        if maj == 0:
            ax.text(2007, 87, "Men")
            ax.text(2005, 10, "Women")
        if maj == 5:
            ax.text(2007, 63, "Men")
            ax.text(2005, 35, "Women")
            ax.tick_params(labelbottom = "on")

for maj in range(0, 5):
        ax = fig.add_subplot(6, 3, ((maj*3)+2))
        ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[maj]], c = cb_dark_blue, label = "Women", linewidth =3)
        ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[maj]], c = cb_orange, label = "Men", linewidth = 3)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0, 100)
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "off")
        ax.set_title(lib_arts_cats[maj])
        ax.set_yticks([0,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        for spot in spines:
            ax.spines[spot].set_visible(False)
        if maj == 0:
            ax.text(2005, 73, "Women")
            ax.text(2006, 22, "Men")
        if maj == 4:
            ax.text(1970, 67, "Men")
            ax.text(1970, 28, "Women")
            ax.tick_params(labelbottom = "on")

for maj in range(0, 6):
        ax = fig.add_subplot(6, 3, ((maj*3)+3))
        ax.plot(women_degrees["Year"], women_degrees[other_cats[maj]], c = cb_dark_blue, label="Women", linewidth =3)
        ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[maj]], c = cb_orange, label = "Men", linewidth = 3)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0, 100)
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom = "off")
        ax.set_title(other_cats[maj])
        ax.set_yticks([0,100])
        ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
        for spot in spines:
            ax.spines[spot].set_visible(False)
        if maj == 0:
            ax.text(2005, 88, "Women")
            ax.text(2007, 7, "Men")
        if maj == 5:
            ax.text(2007, 63, "Men")
            ax.text(2005, 34, "Women")
            ax.tick_params(labelbottom = "on")
fig.suptitle('Visualizing The Gender Gap In College Degrees(1968 - 2011) By FADLULLAH ADEOLA RAJI (PREDSTAN)', fontsize=18)
fig.savefig("gender_degrees.png")
plt.show()

```
